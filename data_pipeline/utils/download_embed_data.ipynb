```python
import pandas as pd

df = pd.read_json("hf://datasets/anti-ai/ViNLI-Zalo-supervised/law_vi.jsonl.gz", lines=True)
```

```python
df
```

```python
import os

# To th mc data/embed nu cha tn ti
embed_dir = "../../data/embed"
os.makedirs(embed_dir, exist_ok=True)

print(f"Th mc {embed_dir}  c to thnh cng!")
print(f"Dataset c {len(df)} dng d liu")
```

```python
# Lu dataset di nhiu nh dng khc nhau trong th mc data/embed

# 1. Lu di dng JSON Lines (gi nguyn nh dng gc)
jsonl_path = os.path.join(embed_dir, "law_vi.jsonl")
df.to_json(jsonl_path, orient='records', lines=True, force_ascii=False)
print(f"  lu dataset di dng JSONL: {jsonl_path}")


print(f"\n Tng kt:")
print(f"- S lng bn ghi: {len(df):,}")
print(f"- S ct: {len(df.columns)}")
print(f"- Ct c sn: {list(df.columns)}")
print(f"- Th mc lu tr: {os.path.abspath(embed_dir)}")
```

```python
# Kim tra ni dung ca dataset  m bo d liu  c ti chnh xc
print(" Xem trc d liu:")
print(f"Shape: {df.shape}")
print("\nCc ct:")
for col in df.columns:
    print(f"- {col}: {df[col].dtype}")

print("\n Mu d liu u tin:")
print("="*50)
for col in df.columns:
    if len(df) > 0:
        sample_value = str(df[col].iloc[0])
        if len(sample_value) > 100:
            sample_value = sample_value[:100] + "..."
        print(f"{col}: {sample_value}")
print("="*50)
```
